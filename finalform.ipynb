```python
import pandas as pd
import numpy as np
```

```python
#load data from csv
df = pd.read_csv('public/data/pivot_df.csv')
```

```python
df.head()
```

```python

# Melt-Funktion verwenden, um den DataFrame umzustrukturieren
melted_df = pd.melt(df, id_vars=['Country', 'City', 'Latitude', 'Longitude'], var_name='Date', value_name='Temperature')

# Sortieren nach Country, City und Date
melted_df = melted_df.sort_values(by=['Country', 'City', 'Date'])

# Vor dem Gruppieren die Temperaturdaten in Arrays umwandeln
melted_df['Temperature'] = melted_df['Temperature'].apply(lambda x: [x] if pd.notna(x) else ["x"])

# Gruppieren nach Country und City und die Temperaturen als Liste speichern
grouped_df = melted_df.groupby(['Country', 'City', 'Latitude', 'Longitude'])['Temperature'].sum()

# DataFrame wieder zusammenfügen
result_df = grouped_df.reset_index()
```

```python
result_df.head()
```

```python
result_df.to_csv('public/data/pivot_df_melted.csv', index=False)
```
